```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

    def __str__(self):
        string = ""
        temp = self.head
        while temp is not None:
            if temp.next is None:
                string += str(temp.value)
            else:
                string += str(temp.value) + " => " 
            temp = temp.next
        return string

l1 = LinkedList()
l1.append(1)
l1.append(4)
l1.append(3)
l1.append(2)
l1.append(5)
l1.append(2)
print(l1)

```
