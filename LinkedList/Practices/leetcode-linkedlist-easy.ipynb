```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        
class LinkedList:
    def __init__(self, value=None):
        if value is None:
            self.head = None
            self.tail = None
            self.length = 0
        else:
            new_node = Node(value)
            self.head = new_node
            self.tail = new_node
            self.length = 1
            
    def __str__(self):
        if self.head is None:
            return "None"
        temp = self.head
        string = ""
        while temp:
            if temp.next is None:
                string += f"{temp.value}"
            else:
                string += f"{temp.value} => "
            temp = temp.next
        return string
            
    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1
        
    def prepend(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self.length += 1
        
    def pop(self):
        if self.head is None:
            return None
        pre = self.head
        temp = self.head
        while temp.next:
            pre = temp
            temp = temp.next
        self.tail = pre
        self.tail.next = None
        self.length -= 1
        temp = None
        
    def pop_first(self):
        if self.head is None:
            return None
        temp = self.head
        self.head = self.head.next
        temp = None
        self.length -= 1
        
    def get(self, index):
        if index < 0  or index >= self.length:
            return None
        temp = self.head
        for _ in range(index):
            temp = temp.next
        print(temp.value)
        
    def set_value(self, index, value):
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        for _ in range(index):
            temp = temp.next
        temp.value = value
        
    def remove(self, index):
        if index < 0 or index >= self.length:
            return None
        if index == 0:
            return self.pop_first()
        if index == self.length-1:
            return self.pop()
        pre = self.head
        temp = self.head
        for _ in range(index):
            pre = temp
            temp = temp.next
        pre.next = temp.next
        temp = None
        self.length -= 1
        
    def insert(self, index, value):
        if index < 0 or index >= self.length:
            return None
        new_node = Node(value)
        if index == 0:
            return self.prepend(value)
        pre = self.head
        curr = self.head
        for _ in range(index):
            pre = curr
            curr = curr.next
        pre.next = new_node
        new_node.next = curr
        self.length += 1

    def reverse(self):
        pre = None
        curr = self.head
        while curr:
            next = curr.next
            curr.next = pre
            pre = curr
            curr = next
        self.head = pre
    
l1 = LinkedList()
l1.append(1)
l1.append(20)
l1.append(30)
l1.append(40)
l1.append(50)
# l1.reverse()
print(l1)
```

```python
# easy
# https://leetcode.com/problems/middle-of-the-linked-list/description/

l3 = LinkedList()
l3.append(10)
l3.append(20)
l3.append(30)
l3.append(40)
l3.append(50)

head = l3.head

def find_middle_node(head):
    if head is None:
        return None
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow.value

print(find_middle_node(head))
```

```python
# easy
# https://leetcode.com/problems/reverse-linked-list/description/

l4 = LinkedList()
l4.append(10)
l4.append(20)
l4.append(30)
l4.append(40)
l4.append(50)

head = l4.head

# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

def reverseList(head):
    """
    :type head: ListNode
    :rtype: ListNode
    """
    pre = None
    curr = head
    while curr:
        next = curr.next
        curr.next = pre
        pre = curr
        curr = next
    return pre
    
    
reverseLinkedList = reverseList(head)
print(reverseLinkedList.value)
```

```python
# easy
# https://leetcode.com/problems/merge-two-sorted-lists/description/

# Input: list1 = [1,2,4], list2 = [1,3,4]
# Output: [1,1,2,3,4,4]
# Example 2:

# Input: list1 = [], list2 = []
# Output: []
# Example 3:

# Input: list1 = [], list2 = [0]
# Output: [0]

l1 = LinkedList()
l1.append(1)
l1.append(2)
l1.append(4)
l2 = LinkedList()
l2.append(1)
l2.append(3)
l2.append(3)
l2.append(4)

print(l1)
print(l2)

l1head = l1.head
l2head = l2.head


def mergeTwoLists(list1, list2):
    
    dummy = cur = Node(0)
        
    if list1 is None:
        return list2
    if list2 is None:
        return list1
    
    while list1 and list2:
        if list1.value < list2.value:
            cur.next = list1
            list1 = list1.next
        else:
            cur.next = list2
            list2 = list2.next
        cur = cur.next
    cur.next = list1 or list2
    return dummy.next
    
    
output = mergeTwoLists(l1head, l2head)

temp = output
while temp:
    print(temp.value)
    temp = temp.next
    
```

```python
# easy 
# https://leetcode.com/problems/palindrome-linked-list/description/

l1 = LinkedList(1)
l1.append(2)
l1.append(2)
l1.append(1)

print(l1)



def isPalindrome(head):
    """
    :type head: ListNode
    :rtype: bool
    """
    checker_list = []
    temp = head
    while temp:
        checker_list.append(temp.value)
        temp = temp.next
    if checker_list == checker_list[::-1]:
        return True
    return False
    
    
    
    
print(isPalindrome(l1.head))
    
        
```

```python
# easy
# https://leetcode.com/problems/linked-list-cycle/description/

l1 = LinkedList(1)
l1.append(2)
l1.append(3)
l1.append(4)

print(l1)
print('---')

def makeCycle(head, index):
    temp = head
    while temp.next is not None:
        temp = temp.next
    tail = temp
    
    temp_two = head
    for _ in range(index):
        temp_two = temp_two.next
    
    # make it a cycle
    tail.next = temp_two
    
#     checker = head
#     for _ in range(10):
#         print(checker.value)
#         checker = checker.next

def hasCycle(head):
    """
    :type head: ListNode
    :rtype: bool
    """
    if head is None:
        return None
    slow = head
    fast = head
    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
        
    

makeCycle(l1.head, 1)

print(hasCycle(l1.head))
```

```python
# easy
# https://leetcode.com/problems/remove-linked-list-elements/
# Input: head = [1,2,6,3,4,5,6], val = 6
# Output: [1,2,3,4,5]
l1 = LinkedList(1)
l1.append(2)
l1.append(6)
l1.append(3)
l1.append(4)
l1.append(5)
l1.append(6)

l2 = LinkedList(7)
l2.append(7)
l2.append(7)

print(l2)
print('---')

def removeElements(head, val):
    """
    :type head: ListNode
    :type val: int
    :rtype: ListNode
    """
    if head is None:
        return None

    dummy = Node(0)
    dummy.next = head
    curr = dummy
    while curr.next:
        if curr.next.value == val:
            curr.next = curr.next.next
        else:
            curr = curr.next
    return dummy.next

    


print(removeElements(l2.head, 7))

print(l2)

    
```
