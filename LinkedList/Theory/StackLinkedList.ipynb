```python
class StackNode:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class StackLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def print_list(self):
        if self.head is None:
            return None
        sll = []
        temp = self.head
        while temp is not None:
            sll.append(temp.value)
            temp = temp.next
        print(sll)

    def push(self, value):
        new_node = StackNode(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

    def pop(self):
        if self.head is None:
            return None
        pre = self.head
        temp = self.head
        while temp.next is not None:
            pre = temp
            temp = temp.next
        self.tail = pre
        self.tail.next = None
        self.length -= 1
        if self.length == 0:
            self.head = None
            self.tail = None

    def top(self):
        print(self.tail.value)
        return

    def isEmpty(self):
        if self.length == 0:
            return True
        return False

    def size(self):
        print(self.length)
        return self.length

    
```

```python
sll = StackLinkedList()
sll.push(10)
sll.push(20)
sll.push(30)
sll.pop()
sll.top()
sll.size()
sll.print_list()
```
