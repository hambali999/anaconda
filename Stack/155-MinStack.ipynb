```python
class MinStack:
    def __init__(self):
        self.list = []
        self.length = 0
        
    def push(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.list.append(val)
        self.length += 1
        
    def pop(self):
        """
        :rtype: None
        """
        self.list.pop()

    def top(self):
        """
        :rtype: int
        """
        return self.list[-1]

    def getMin(self):
        """
        :rtype: int
        """
        arr = [x for x in self.list]
        minimum = min(arr)
        return minimum
        
        


# Input
# ["MinStack","push","push","push","getMin","pop","top","getMin"]
# [[],[-2],[0],[-3],[],[],[],[]]

# Output
# [null,null,null,null,-3,null,0,-2]

# Explanation
minStack = MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.pop()
minStack.getMin(); 
print(minStack.getMin())
# minStack.pop();
# minStack.top();    // return 0
# minStack.getMin(); // return -2
```
