```python
'''

Depth first search:
- Preorder traversal
    = Preorder traversal visits the node in the order: Root -> Left -> Right

- Inorder traversal
    = 
    
- Post order traversal

Breadth first search:
- Level order traversal

'''
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
        
newbt = TreeNode(1)
newbtleft = TreeNode(2)
newbtright = TreeNode(3)
newbt.left = newbtleft
newbt.right = newbtright
newbtleft.left = TreeNode(4)
newbtleft.right = TreeNode(5)
newbtright.left = TreeNode(6)
newbtright.right = TreeNode(7)
```

```python
def pre_order_traversal(rootNode):
    if not rootNode:
        return
    print(rootNode.data)
    pre_order_traversal(rootNode.left)
    pre_order_traversal(rootNode.right)



print(f"=== Preorder Traversal of Binary Tree ===\n")
pre_order_traversal(newbt)
print('---'*20)
```

```python
def in_order_traversal(rootnode):
    if not rootnode:
        return
    in_order_traversal(rootnode.left)
    print(rootnode.data)
    in_order_traversal(rootnode.right)

print(f"=== Inorder Traversal of Binary Tree ===\n")
in_order_traversal(newbt)
print('---'*20)
```

```python
def post_order_traversal(rootnode):
    if not rootnode:
        return
    post_order_traversal(rootnode.left)
    post_order_traversal(rootnode.right)
    print(rootnode.data)

print(f"=== Postorder Traversal of Binary Tree ===\n")
post_order_traversal(newbt)
print('---'*20)
```

```python
def level_order_traversal(rootnode):
    if rootnode is None:
        return None
    
    queue = []
    queue.append(rootnode)
    
    while len(queue) > 0:
        print(queue[0].data)
        node = queue.pop(0)
        
        if node.left:
            queue.append(node.left)
            
        if node.right:
            queue.append(node.right)
    

print(f"=== Level order Traversal of Binary Tree ===\n")
level_order_traversal(newbt)
print('---'*20)
```

```python
# search - level order traversal

def search_level_order_traversal(rootnode, value):
    if not rootnode:
        return
    if rootnode.data == value:
        return True
    queue = []
    queue.append(rootnode)
    while len(queue) > 0:
        node = queue.pop(0)
        if node.data == value:
            return True
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return False
        



print(search_level_order_traversal(newbt, 8))
```

```python
# insert - level order traversal

newbinarytree = TreeNode(1)
newbtleft = TreeNode(2)
newbtright = TreeNode(3)
newbinarytree.left = newbtleft
newbinarytree.right = newbtright
# newbtleft.left = TreeNode(4)
newbtleft.right = TreeNode(5)
newbtright.left = TreeNode(6)
newbtright.right = TreeNode(7)

def insert_level_order_traversal(rootnode, value):
    if rootnode is None:
        rootnode = TreeNode(value)
        return rootnode
    queue = []
    queue.append(rootnode)
    while len(queue) > 0:
        node = queue.pop(0)
        if node.left is None:
            node.left = TreeNode(value)
            break
        else:
            queue.append(node.left)
        if node.right is None:
            node.right = TreeNode(value)
            break
        else:
            queue.append(node.right)

insert_level_order_traversal(newbinarytree, 10)

level_order_traversal(newbinarytree)
```
