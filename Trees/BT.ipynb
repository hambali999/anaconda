```python
'''

Depth first search:
- Preorder traversal
    = The root node of the subtree is visited first.
    = Then the left subtree  is traversed.
    = At last, the right subtree is traversed.
- Inorder traversal
    = 
- Post order traversal

Breadth first search:
- Level order traversal

'''
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.leftChild = None
        self.rightChild = None
        
        
newBT = TreeNode("Drinks")
leftChild = TreeNode("Hot")
rightChild = TreeNode("Cold")
newBT.leftChild = leftChild
newBT.rightChild = rightChild
rightChild.rightChild = TreeNode("Ice Milk Tea")
rightChild.leftChild = TreeNode("Ice Lemon Tea")
leftChild.leftChild = TreeNode("Hot Milo")
leftChild.rightChild = TreeNode("Hot Tea")
# leftChild.leftChild.leftChild = TreeNode("Hot Choc Milo")


def preOrderTraversal(rootNode, depth=0):
    if not rootNode:
        return
    print(' ' * 4 * depth + '|__' + str(rootNode.data))
    preOrderTraversal(rootNode.leftChild, depth + 1)
    preOrderTraversal(rootNode.rightChild, depth + 1)
    

        
def inOrderTraversal(rootNode, depth=0):
    if not rootNode:
        return
    inOrderTraversal(rootNode.leftChild, depth + 1)
    print(' ' * 4 * depth + '|__' + str(rootNode.data))
    inOrderTraversal(rootNode.rightChild, depth + 1)
    
def postOrderTraversal(rootNode, depth=0):
    if not rootNode:
        return
    postOrderTraversal(rootNode.leftChild, depth + 1)
    postOrderTraversal(rootNode.rightChild, depth + 1)
    print(' ' * 4 * depth + '|__' + str(rootNode.data))
    
# print(f"=== Preorder Traversal of Binary Tree ===\n")
# preOrderTraversal(newBT)
# print('---'*20)
print(f"=== Inorder Traversal of Binary Tree ===\n")
inOrderTraversal(newBT)
print('---'*20)
print(f"=== Postorder Traversal of Binary Tree===\n")
postOrderTraversal(newBT)
```
